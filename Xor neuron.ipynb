Цель нейронной сети — классифицировать объект в соответствии с приведенной выше таблицей истинности. Если построить входы в соответствии с их выходами, видно, что эти точки не являются линейно разделимыми.


Нейронная сеть должна создавать две разные разделяющие плоскости для разделения входных данных. Это достигается за счет использования скрытых слоев (hidden layers). Нейронная сеть будет состоять из одного входного слоя с двумя нейронами (X1,X2); один скрытый слой с двумя нейронами (поскольку необходимы две плоскости); и один выходной слой (Y). Таким образом, нейронная сеть выглядит так:

1. Инициализируйте веса(weights) и смещения(bias) случайным образом.
2.      Итерация по данным
     1. Вычислите прогнозируемый результат с помощью сигмоиды
     2. Вычислите потери, используя функцию квадратичной ошибки
     3. W(новый) = W(старый) — α ∆W
     4. B(новый) = B(старый) — α ∆B
3. Повторяйте, пока ошибка не станет минимальной
