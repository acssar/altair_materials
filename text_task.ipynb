```python
import codecs
import numpy as np
import gensim
import re
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
# with codecs.open('data/RedOnRed.fb2', encoding='utf-8', mode='r') as f:
with codecs.open('/home/roman/Downloads/RedOnRed.fb2', encoding='utf-8', mode='r') as f:    
    docs = f.readlines()
    
max_sentence_len = 12
sentences = [sent for doc in docs for sent in doc.split('.')]
sentences = [[re.sub('[^а-яА-Я%]', ' ', word) for word in sent.lower().split()[:max_sentence_len]] for sent in sentences]
print(len(sentences), 'предложений')

```

```python
sentences
```

```python
word_model = gensim.models.Word2Vec(sentences, vector_size=100, min_count=1, window=5, epochs=100)
```

```python
pretrained_weights = word_model.wv.vectors
vocab_size, embedding_size = pretrained_weights.shape
print(vocab_size, embedding_size)
```

```python
print('Похожие слова:')
for word in ['трон', 'дом', 'конь', 'дуэль', 'выстрел']:
    most_similar = ', '.join('%s (%.2f)' % (similar, dist) for similar, dist in word_model.wv.most_similar(word)[:8])
    print('  %s -> %s' % (word, most_similar))

```
